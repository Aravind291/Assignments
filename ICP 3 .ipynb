```python
import pandas as pd
import numpy as np
df=pd.read_csv('data.csv')
df.describe() #Basic statistical description
```

Uploaded and read the given file "data.csv"

```python
df.isnull().sum() #checking for null values
```

Checking the given data set for null values.

```python
df['Calories'].fillna(df['Calories'].mean(),inplace=True)# replacing the null values with mean values
df['Calories'].isnull().sum() #checking if null value still exists
```

checking and replacing null values with mean values

```python
df.groupby(['Duration','Pulse']).agg({'Calories':['min','max','count','mean'],'Maxpulse':['min','max','count','mean']})
```

Selected the columns "Duration" and "Maxpulse" and the aggregate for the "minimum" , "maximum", "count" and "mean" is displayed.

```python
df[(df['Calories'].between(500,1000))]#calories between 500 and 1000
```

Rows with Calorie values between 500 and 1000 is displayed.


```python
df[(df['Calories'] > 500) & (df['Pulse'] <= 100)]#calories >500 and pulse<100 data
```

Rows with calorie values greater than 500 and pulse value more than 100 is displayed.

```python
df_modified=df.loc[:,df.columns!='Maxpulse']
df_modified#df without maxpulse
```

Created "df_modified" data frame excluding "Maxpulse" column.

```python
df.drop('Maxpulse',axis=1) #deleting Maxpulse in the main df
```

"Maxpulse" column has been deleted.

```python
 df['Calories'] = df['Calories'].fillna(0).replace([np.inf, -np.inf], 100000).astype(int)
print("\nDataframe after converting 'Calories' to int:")
print(df.dtypes)
```

Converted the data type of "Calories" to int.

```python
df.plot.scatter(x='Duration',y='Calories') #scatter plot
```

Used "pandas" to create a scatterplot for the columns "Duration" and " Calories"

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(sdf.iloc[:, :-1].values,sdf.iloc[:,1].values,test_size =0.2)
x_train#checking train data
```

Checking the train data from the given dataset.

```python
from sklearn.linear_model import LinearRegression
m=LinearRegression()#linearregression
m.fit(x_train, y_train)#fitting the data for the linear regression
```

```python
y_pred=m.predict(x_test)#predicting the data
#y_pred*z+min(sdf['Salary'])
```

```python
import math
from sklearn.metrics import mean_squared_error as ms
ms(y_pred,y_test)#mean square error
```
